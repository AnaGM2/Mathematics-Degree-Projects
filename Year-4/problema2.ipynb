## Problema 2 
Si disponemos de una muestra $Y_1,...,Y_n$ de variables aleatorias independientes con la misma distribución, indica cuáles de los siguientes modelos se trata de un Modelo Lineal Generalizado:

- $E(Y_i|x_i) = (\beta_0 + \beta_1x_i)^2$ y cada $Y_i$ sigue una distribución de Pareto.
- $E(Y_i|x_i) = \mu_i = \beta_0+ log(\beta_1 + \beta_2x_i); Y_i \sim N(\mu_i,\sigma^2)$


### Solución:


- Por el Problema 1 sabemos que la función distribución de pareto es $ f(y,\theta) = \theta y^{-\theta -1}$, la cual se puede escribir de la forma $f(y, \theta) = \exp\left(a(y)b(\theta) + c(\theta) + d(y)\right)$ con 
  $$a(y) = \log(y)$$
  $$b(\theta) = -\theta - 1$$$$c(\theta) = \log \theta$$
  $$d(y) = 0$$
  por lo que pertenece a la familia exponencial. Por tanto, tenemos que las $Y_1,...,Y_n$ siguen una distribución de la familia exponencial, pero como 
 $a(y) = \log(y)$, tenemos que la distribución no tiene la forma canónica. Por tanto, no se trata de un Modelo Lineal Generalizado.
 
 
- Como $Y_i$ sigue una distribución normal, tenemos que $Y_1,...,Y_n$ siguen una distribución de la familia exponencial con forma canónica.
Buscamos ahora la relación entre el predictor lineal $\nu_i = x_i\beta$ y la media $\mu_i = E[Y_i]$ a través de la una función g monótona y diferenciable (función link): 
$$g(\mu_i) = x_i^t\beta$$


Si tomamos la función exponencial a los dos lados:
$$
\mu_i=\beta_0+\log(\beta_1+\beta_2x_i) \Rightarrow e^{\mu_i} = e^{\beta_0}\left( \beta_1+\beta_2x_i \right)
$$ 
Vemos que en la expresión de la derecha en ningún caso se puede despejar $x_i^T\beta$ e igualarlo a una función de $\mu_i$, por tanto, no se trata de un Modelo Lineal Generalizado.
