1. **Prueba que si** $p$ **es primo impar,** $2(p-3)! \equiv -1 \text{ mod } p$ 



Usaremos el Teorema de Wilson, que afirma que dado $p$ primo se cumple que $(p-1)! \equiv -1 \text{ mod }p$ 

Tengamos en cuenta que $(p-1) \equiv_p -1$, $(p-2) \equiv_p -2$ y que $(p-1)! = (p-1)(p-2)(p-3)!$ . Con esta información, obtenemos:

$-1 \equiv_p (p-1)! \equiv_p (p-1)(p-2)(p-3)! \equiv_p (-1)(-2)(p-3)! \equiv_p 2(p-3)!$ 

Por tanto, podemos concluir que $2(p-3)! \equiv -1 \text{ mod } p$ 


**2. Muestra que \-1 es residuo cuadrático para los primos de la forma 4k\+1y no lo es para los de la forma 4k\+3.**


Por el lema de Euler sabemos que $(\frac{a}{p})=a^{\frac{p-1}{2}} \ mod \ p$ . Sea entonces $a=-1$ y $p=4k+1$ tenemos:
\begin{equation*}
\left(\frac{-1}{4k+1}\right) =(-1)^{\frac{4k+1-1}{2}}=(-1)^{\frac{4k}{2}}=(-1)^{2k}=1
\end{equation*}
Por tanto, $-1$ si es residuo cuadrático para los primos de la forma $4k+1$.<br>
Volvemos a usar el lema de Euler para los primos de la forma $4k+3$ y obtenemos lo siguiente:
\begin{equation*}
\left(\frac{-1}{4k+3}\right) =(-1)^{\frac{4k+3-1}{2}}=(-1)^{\frac{4k+2}{2}}=(-1)^{2k+1}=-1
\end{equation*}
Por lo tanto, $-1$ no es residuo cuadrático para los primos de la forma $4k+3$.

4. **Encuentra los residuos cuadráticos** $\text{ mod } 29$ 



Los residuos cuadráticos del primo $p$  son $1, 2^2, 3^2, \dots, \left(\dfrac{p-1}{2}\right)^2$ . Los calcularemos con python y teniendo en cuenta que $\frac{p-1}{2} = 14$  



```python
def modulo(numero, modulo):
    resto = numero//modulo
    numero_mod = numero - resto*modulo
    return numero_mod

residuos = []
for i in range(1, 15):
    residuos.append(modulo(i**2, 29))
residuos.sort()
    
print('Los residuos cuadráticos módulo 29 son: ', residuos)
```

Así, el teorema Chino de los restos nos daría una solución única módulo $pq = n$ para cada sistema, que será solución de $x^2 \equiv a$ mod $n$, de forma que en total obtendríamos 4 soluciones de $x^2 \equiv a$ mod $n$.

$x \equiv -1753 \cdot 3823 \cdot 2472 - 1425 \cdot 3803 \cdot 1338$ mod $14538869 \equiv 11566973 = (21,20,17,2411)_{27} =$ UTQXK

Los 4 posibles mensajes son EOZRF, GAUSS, TGVVY y UTQXK, por lo que Alice elige GAUSS.

```python
mp = 1753
mq = 1425

def euclides(a, b):
    
    if b == 0:
        return (a, 1, 0)
    
    else:
        q = a // b
        r = a % b
        (mcd, u0, v0) = euclides(b, r)
        u = v0
        v = u0-q*v0
        return (mcd, u, v)

def inversa(a,m):
    v = euclides(a,m)[1]
    return v % m

y1 = inversa(q,p)
y2 = inversa(p,q)

x1 = (mp*q*y1 + mq*p*y2) % n
x2 = (-mp*q*y1 + mq*p*y2) % n
x3 = (mp*q*y1 - mq*p*y2) % n
x4 = (-mp*q*y1 - mq*p*y2) % n

x1, x2, x3, x4
```

**23.  Imagina que Miguel tiene una amante, Marta, con la que quiere intercambiar mensajes secretos usando el sistema ElGamal. Imagina que Miguel usa el mismo exponente secreto para mandar dos mensajes $M_1$ y $M_2$ a Marta. ¿Qué sucede si su novia Olga intercepta los mensajes? ¿Los puede leer? ¿Qué pasa si se imagina lo que hay en uno de ellos? ¿Consigue leer el otro mensaje?**


Supongamos que Miguel envía dos mensajes $m$ y $\bar{m}$ con el mismo b, entonces el oyente obtendrá
\begin{equation*}
c \equiv A^b m \ mod \ p \ y \ \bar{c} \equiv A^b \bar{m} \ mod \ p
\end{equation*}

De aquí podemos sacar por tanto que $\bar{c} c^{-1} \equiv \bar{m} m^{-1} \ mod \ p$. Es decir que si Olga no conoce ninguno de los mensajes no va a poder descifrarlos, pero si Olga se imagina lo que pone en uno de los dos mensajes entonces si podrá leer el otro mensaje, ya que supongamos que conoce el contenido de m, entonces
\begin{equation*}
\bar{m} \equiv \bar{c} c^{-1} m \ mod \ p
\end{equation*}
y como conocemos todos los términos de la derecha podemos calcular $\bar{m}$, es decir, descifrar el otro mensaje.



```python
from math import trunc,log
def num_posiciones(num, len_alfabeto):
    k = trunc(log(num,len_alfabeto))
    posiciones =  []
    contador = k
    for i in range(0,k):
        coc = num//(len_alfabeto**(contador))
        resto = num%(len_alfabeto**(contador))
        num = resto
        contador -= 1
        posiciones.append(coc)
    posiciones.append(resto)
    return posiciones

print(num_posiciones(441,27))
    
    
        
    
```
