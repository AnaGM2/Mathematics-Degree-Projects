1. **Estime el tiempo para calcular** $n^m$**, con** $n,m\in \mathbb{N}$**.**



Tenemos que hacer $m-1$ multiplicaciones de $2$ números de tamaño $\leq n$ uno y de tamaño $\leq n^m$ el otro.

$n\cdot n\cdot n \cdots n = n^2\cdot n \cdots n = n^3\cdots n = \dots = n^m$

Por tanto, la complejidad del algoritmo es $\theta((m-1)\log_2 (n) \log_2 (n^m)) = \theta(m^2 \log_2^2(n))$.

El algoritmo es de tiempo polinomial en $n$ y de tiempo exponencial en $m$.

3. **Escriba** $(17,37)$ **como combinación lineal de** $17$ **y** $37$**.** 



Tomamos $a=37$ y $b=17$ y aplicamos el Algortimo de Euclides:

$37 = 2\cdot 17 + 3 \Rightarrow q_0 = 2, r_0 = 3$

$17 = 5\cdot 3 + 2 \Rightarrow q_1 = 5, r_1 = 2$

$3 = 1\cdot 2 + 1 \Rightarrow q_2 = 1, r_2 = 1$

$2 = 2\cdot 1 \Rightarrow q_3 = 2, r_3 = 0$

Por tanto, $(17,37) = r_2 = 1$. Además:

$37 = 2\cdot 17 + 3 \Rightarrow 3 = 37 - 2\cdot 17$

$17 = 5\cdot 3 + 2 \Rightarrow 2 = 17 - 5\cdot 3 = 17 - 5 (37 - 2\cdot 17) = -5\cdot 37 + 11\cdot 17$

$3 = 1\cdot 2 + 1 \Rightarrow 1 = 3 - 1\cdot 2 = 37 - 2\cdot 17 - 1 (-5\cdot 37 + 11\cdot 17) = 6\cdot 37 - 13\cdot 17$

$-13\cdot 17 +  6\cdot 37 = (17,37)$

También podemos obtener este resultado usando la siguiente función:

```python
def euclides(a, b):
    
    if b == 0:
        return (a, 1, 0)
    
    else:
        q = a // b
        r = a % b
        (mcd, u0, v0) = euclides(b, r)
        u = v0
        v = u0-q*v0
        return (mcd, u, v)
    
euclides(37,17)
```

El primer elemento que devuelve es $(37,17) = (17,37) = 1$, mientras que el segundo y el tercer elemento son los coeficientes de la combinación lineal buscada.

8. **Muestra que existen infinitos primos de la forma 4n\+3**



Procedemos por reducción al absurdo, para ello suponemos que existe una cantidad finita de primos $(p_1,p_2,...,p_q)$ de la forma $4n+3$. <br>
Tomamos ahora $N=4p_1 p_2\cdot \cdot \cdot p_q-1=4(4n_1+3)(4n_2+3)\cdot \cdot \cdot (4n_q+3)-1=4((4n_1+3)(4n_2+3)\cdot \cdot \cdot (4n_q+3)-1)+3$ por lo tanto N también es de la forma 4n+3.<br>
Primero, si suponemos que N es primo, debe estar en $\{p_1,p_2,...,p_q\}$ pero esto no puede ser porque $N>p_i \forall i \in \{1,2,...,q\}$, por lo tanto habríamos encontrado otro primo distinto de la forma 4n+3, por lo que llegaríamos a una contradicción.<br>

En segundo lugar, suponemos que N no es primo, entonces N tendrá una factorización en números primos, es decir, $N=r_1,r_2\cdot \cdot \cdot r_t$ donde los $r_i$ son primos no necesariamente distintos entre sí. Sabemos que N es de la forma 4n+3, por lo tanto es impar y eso significa que todos los factores también lo serán, es decir, que serán de la forma 4n+1 o 4n+3.<br>

Vemos ahora que si multiplicamos dos números de la forma 4n+1 el resultado es de la misma forma: $(4n+1)(4m+1)=4(4nm+n+m)+1$. Esto implica que si todos los $r_i$ fueran de la forma 4n+1, N también lo sería y sabemos que N no lo es, por tanto debe haber algunos $j\in\{1,2,...,t\}, i\in\{1,2,...,q\}$ tal que $r_j=p_i$. Como $r_j$ es uno de los factores primos de N, $r_j$ divide a N y por tanto $p_i$ divide a N, pero por la construcción de N tenemos que $N+1=4p_1p_2 \cdot \cdot \cdot p_q$ por tanto $p_i$ divide también a N+1, esto implica que $p_i \ | \ (N+1)-N \Rightarrow p_i \ | \ 1 \Rightarrow p_i=1$ pero esto es una contradicción porque $p_i$ es primo.<br>

En ambos supuestos hemos llegado a una contradicción que viene de suponer que existe una cantidad finita de primos de la forma 4n+3, por lo que podemos concluir que existe una cantidad infinita de primos de la forma 4n+3


9. **Sean** $p$ **y** $q$ **dos primos impares consecutivos. Muestre que cualquier factorización de** $p + q$ **en primos necesita de por lo menos tres primos no necesariamente distintos. Por ejemplo** $5 + 7 = 2 \times 2 \times 3$**.**



Por ser $p$ y $q$ impares consecutivos, existe $k \in\mathbb{Z}$ tal que:

$p = 2k + 1$

$q = 2k + 3$

y por ser primos, necesariamente $k \in\mathbb{N}$, pues los menores primos impares consecutivos son $3$ y $5$, luego:

$2k+1 \geq 3 \Leftrightarrow 2k\geq 2 \Leftrightarrow k \geq 1$

Entonces, la suma queda:

$p+q = (2k+1) + (2k+3) = 4k+4 = 4(k+1) = 2\cdot 2 (k+1)$

Luego cualquier factorización de $p+q$ en primos necesita de al menos el $2$ repetido dos veces, junto con la factorización de $k+1 \geq 2$, que tendrá por lo menos un primo.

Es decir, la factorización de $p+q$ en primos estará compuesta por al menos tres factores primos no necesariamente distintos.

**ITFMGWTITJWARGIJGTRWGTFGLIPNIWJGFWLWABGO**

**KWFWLGJWGGTFGLIPGCWAMJZMWJKTEGPBGOVKGT**

**FGLIPYGLEGAOKCKLGOKCMFGNJMGYFMDWFW**

**LMOGFGCWTRWGLCKPKFWLGOKPMTGSIW**

**SIWJMGSIWCWLGAMJZMWJGTOGLMWTRWYN**

**MTGLCWTRWOMTAWEIMOKCMFGOGLMWTRW**


```python
def analisis_frecuencia(ciphertext):
    ciphertext = ciphertext.upper()
    
    frecuencias = {}
    for caracter in ciphertext:
        if caracter in frecuencias:
            frecuencias[caracter] += 1
        else:
            frecuencias[caracter] = 1
    frecuencias_ordenadas = sorted(frecuencias.items(), key=lambda x: x[1], reverse=True)
    
    letra1 = frecuencias_ordenadas[0][0]
    letra2 = frecuencias_ordenadas[1][0]
    
    # A ----> letra1  <==>  0 ----> y1
    # E ----> letra2  <==>  4 ----> y2
    alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    y1 = alfabeto.index(letra1)
    y2 = alfabeto.index(letra2)
            
    return y1, y2


ciphertext = ("ITFMGWTITJWARGIJGTRWGTFGLIPNIWJGFWLWABGOKWFWLGJWGGTFGLIPGCWAMJZMW"
        "JKTEGPBGOVKGTFGLIPYGLEGAOKCKLGOKCMFGNJMGYFMDWFWLMOGFGCWTRWGLCKPKFWLGOKP" 
        "MTGSIWSIWJMGSIWCWLGAMJZMWJGTOGLMWTRWYNMTGLCWTRWOMTAWEIMOKCMFGOGLMWTRW")

analisis_frecuencia(ciphertext)  
```

Observamos que las letras que más se repiten en el texto cifrado son la G y la W. Teniendo en cuenta que las letras más frecuentes en español son la A y la E, podemos suponer que la A del texto plano pasa a ser la G en el encriptado, y la E del texto plano pasa a ser la W. De esta forma, obtenemos el sistema:

$a\cdot 0 + b\equiv 6$ mod $26$

$a\cdot 4 + b\equiv 22$ mod $26$

De aquí obtenemos que $b\equiv 6$ mod $26 \Rightarrow 4a + 6\equiv 22$ mod $26 \Leftrightarrow 4a\equiv 16$ mod $26$. Una solución es $a=4$ y la otra es $a = 4 + \frac{26}{2} = 17$. Como $a=4$ no es coprimo con $26$, pues $(4,26)=2$, esta no puede ser la solución correcta (pues necesitamos que $e(\cdot )$ sea inyectiva). Tomamos $a=17$ y $b=6$, y desciframos con la siguiente función:

```python
def cifra_afin(ciphertext,a,b):
    alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    a_inv = inversa(a,26)
    
    plaintext = ""
    for caracter in ciphertext:
        y = alfabeto.index(caracter)
        x = a_inv*(y-b) % 26
        plaintext = plaintext + alfabeto[x]
    return plaintext


a = 17
b = 6
cifra_afin(ciphertext, a, b)
```

```python
def desencripta(a,b):
    dic={"a":0,"b":0,"c":0,"d":0,"e":0,"f":0,"g":0,"h":0,"i":0,"j":0,"k":0,"l":0,"m":0,"n":0,"o":0,"p":0,"q":0,"r":0,"s":0,"t":0,"u":0,"v":0,"w":0,"x":0,"y":0,"z":0}
    for i in range(0,26):
        c=(a*i+b)//26
        dic[num_to_ABC(i)]=num_to_ABC(a*i+b-c*26)
    return(dic)
```

```python
def invertir_dic(dic):
    dic_invertido = {v: k for k, v in dic.items()}
    return(dic_invertido)
```

```python
def desencripta_texto(texto,dic):
    nuevo_texto=""
    for i in texto:
        nuevo_texto=nuevo_texto+dic[i]
    return(nuevo_texto)
```

```python
def Ic(x):
    x = x.lower()
    m = len(x)
    
    abecedario = 'abcdefghijklmnopqrstuvwxyz'
    frecuencias = {letra: 0 for letra in abecedario}
    for caracter in x:
        frecuencias[caracter] += 1
        
    Ic = 0
    for f in frecuencias.values():
        Ic += f*(f-1)/(m*(m-1))
    return Ic
```

```python
def formar_subtextos(ciphertext, n):
    subtextos = []
    
    for i in range(n):
        y = ciphertext[i::n]
        subtextos.append(y)
    
    return subtextos
```

```python
def MIc(x,y):
    x = x.lower()
    y = y.lower()
    m = len(x)
    t = len(y)
    
    abecedario = 'abcdefghijklmnopqrstuvwxyz'
    frecuenciasx = {letra: 0 for letra in abecedario}
    frecuenciasy = {letra: 0 for letra in abecedario}

    for caracter in x:
        frecuenciasx[caracter] += 1
    for caracter in y:
        frecuenciasy[caracter] += 1
        
    f = list(frecuenciasx.values())
    g = list(frecuenciasy.values())
    
    MIc = 0
    for i in range(26):
        MIc += f[i]*g[i]/(m*t)
    return MIc
```

```python
def desplazamiento(y,g):
    y = y.lower()
    abecedario = 'abcdefghijklmnopqrstuvwxyz'
    y_g = ""
    for caracter in y:
        i = abecedario.index(caracter)
        y_g = y_g + abecedario[(i+g)%26]
    return(y_g)
```

```python
def frecuencias(ciphertext,n=1):
    ciphertext = ciphertext.lower()
    if n == 1:
        abecedario = 'abcdefghijklmnopqrstuvwxyz'
        frecuencias = {letra: 0 for letra in abecedario}
        for caracter in ciphertext:
            frecuencias[caracter] += 1
    else:
        frecuencias = {}
        for i in range(0, len(ciphertext)):
            if len(ciphertext[i:i+n]) == n:
                if ciphertext[i:i+n] in frecuencias:
                    frecuencias[ciphertext[i:i+n]] += 1
                else:
                    frecuencias[ciphertext[i:i+n]] = 1  
        frecuencias = sorted(frecuencias.items(), key=lambda x: x[1], reverse=True)
    return frecuencias
```

```python
def posiciones_grupo(ciphertext,grupo):
    ciphertext = ciphertext.lower()
    grupo = grupo.lower()
    posiciones=[]
    i = 0
    while i < len(ciphertext):
        indice = ciphertext.find(grupo,i)
        if indice == -1:
            break
        posiciones.append(indice)
        i = indice + 1
    return posiciones
```

17. **Sea la expansión b\-ádica de un entero** $n$**,** $n = d_{k-1}b^{k-1} + d_{k-2}b^{k-2} + \cdots + d_1b + d_0$**,** $0\leq d_i < b, d_{k-1}\neq 0$**. Probar que** $k=\lfloor \log_b n \rfloor + 1$**.**



Como $0\leq d_i, \forall i \in \{0, \dots, k-1 \}$ y además, para $i=k-1$ se tiene $d_{k-1} \neq 0$, entonces $d_{k-1} \geq 1$. Entonces:

$b^{k-1} \leq d_{k-1} b^{k-1} \leq d_{k-1} b^{k-1} + d_{k-2} b^{k-2} + \dots + d_1 b + d_0 = n$ 

Por otro lado, como $d_i \leq b-1 \forall i \in \{0,\dots, k-1 \}$:

$n = \sum_{i=0}^{k-1} d_i b^i \leq (b-1) \sum_{i=0}^{k-1} b^i = b^k-1 < b^k$

Juntando ambas desigualdades, llegamos a que:

$b^{k-1} \leq n < b^k \Leftrightarrow (k-1) \log_b b \leq \log_b n < k\log_b b \Leftrightarrow k-1 \leq \log_b n < k \Rightarrow k = \lfloor \log_b n \rfloor + 1$

```python
def num_to_bin5(num):
    bol=False
    if num==0:
        bol=True
        num=1
    bi=np.zeros(5)
    k=math.log2(num)
    k=math.floor(k)
    BIN=np.zeros(k+1)
    for i in range(k,0,-1):
        if (num/(2**i))>=1:
            BIN[k-i]=1
            num=num-(2**i)
    if num==1:
        BIN[k]=1
    if len(BIN)!=5:
        for i in range(len(BIN)):
            bi[4-i]=BIN[i]
    else:
        bi=BIN
    if bol==True:
        bi=np.zeros(5)
    return(bi)
```

19. **Construye un LFSR de grado 4 y periodo 15.**



Para un LFSR de grado $m = 4$, sabemos que el ciclo máximo es de $2^m -1 = 2^4 -1= 15$. Por tanto, estamos buscando un LFSR de grado $4$ con el mayor ciclo posible. Si el polinomio característico del LFSR es:

$z_{m+i} = \sum_{j=0}^{m-1} c_j z_{i+j}$ mod $2$

Entonces para que sea de grado máximo debe tener asociado un polinomio primitivo, donde $z_i$ lo asociamos a $x^i$.

Así, un polinomio primitivo de grado $4$ es $1+x+x^4$, por lo que obtenemos un LFSR cuyo polinomio característico es $z_{i+4} = z_i + z_{i+1}$ mod $2$.

Usamos la siguiente función para probar un ejemplo, y observamos que efectivamente tiene periodo 15:

```python
def LFSR(semilla):
    secuencia = semilla.copy()
    secuencia.append((semilla[0] + semilla[1])%2)
    i=1
    while secuencia[i:] != semilla:
        secuencia.append((secuencia[i] + secuencia[i+1])%2)
        i += 1
    return secuencia
    
    
semilla = [1,0,0,0]
LFSR(semilla)
```

20. **Considere la ecuación**

$a_nx^n + a_{n−1}x^{n−1} + · · · + a_1x + a_0 = 0$

**donde** $a_i$ **es un entero, 0 ≤ i ≤ n. Muestra que si la ecuación tiene una solución racional, digamos** $x =\frac{r}{s}$ **con \(r, s\) = 1, entonces $s|a_n$ y $r|a_0$.**



Si $x=\frac{r}{s}$ es solución, sabemos que <br>
$a_n(\frac{r}{s})^n+a_{n-1}(\frac{r}{s})^{n-1}+...+a_1(\frac{r}{s})+a_0=0$<br>
Multiplicamos toda la expresión por $s^n$ (sabemos que podemos hacerlo porque $s \neq 0$ ya que(r,s)=1) y nos queda entonces<br>
$a_nr^n+a_{n-1}r^{n-1}s+...+a_1rs^{n-1}+a_0s^n=0$<br>
Donde podemos despejar la parte del $a_n$ y nos quedaría<br>
$-a_nr^n=a_{n-1}r^{n-1}s+...+a_1rs^{n-1}+a_0s^n \Leftrightarrow -a_nr^n=s(a_{n-1}r^{n-1}+...+a_1rs^{n-2}+a_0s^{n-1})$<br>
Entonces sabemos que s divide a la parte de la derecha y por tanto también debe dividir a la parte de la izquierda por lo que $s | -a_nr^n$. Pero como (r,s)=1, entonces $(r^n,s)=1$ y por tanto $s | a_n$<br>
Con un procedimiento análogo, volvemos a la expresión<br>
$a_nr^n+a_{n-1}r^{n-1}s+...+a_1rs^{n-1}+a_0s^n=0$<br>
y esta vez despejamos la parte del $a_0$ y nos queda<br>
$a_nr^n+a_{n-1}r^{n-1}s+...+a_1rs^{n-1}=-a_0 s^n \Leftrightarrow r(a_nr^{n-1}+a_{n-1}r^{n-2}s+...+a_1s^{n-1})=-a_0 s^n$<br>
Por tanto, como r divide a la parte de la izquierda también debe dividir a la de la derecha, entonces $r | -a_0 s^n$, pero como (s,r)=1, entonces $(s^n,r)=1$ y por tanto $r | a_0$.
