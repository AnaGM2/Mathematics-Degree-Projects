## **Ejercicio 4.3** 


Aplicar el método del ejercicio anterior para calcular soluciones numéricas del PVI $y’= 4xy^{1/2}, y(0)=1$, para   $0 < x <2$, con una longitud de paso $h=0.1$, $0.05$ y $0.025$. Comparar el resultado con la solución exacta $y(x)=(1+x^2)^2$, ¿converge la solución numérica a la solución exacta cuando h->0?

Dado el método numérico: 
$$ y_{n+2}-y_{n+1}= \frac{h}{12}\big[4f(x_{n+2},y_{n+2})+8 f(x_{n+1},y_{n+1})-f(x_{n},y_{n})\big] $$

Si lo aplicamos al PVI dado $y’= 4xy^{1/2}, y(0)=1$, donde $f(x_{n},y_{n})= 4x_ny_n^{1/2}$ se tiene:

$$ y_{n+2}-y_{n+1}= \frac{h}{12}\big[4\cdot 4\cdot x_{n+2}y_{n+2}^{1/2}+8\cdot 4\cdot x_{n+1}y_{n+1}^{1/2}-4\cdot x_{n}y_{n}^{1/2}\big] $$

Podemos hacer un programa muy sencillo para calcualr las solucines, como unicos parámetros de entrada vamos a considerar el rango de integración y ek número de pasos. Dado que al aplicar el método a este PVI da lugar a una ecuación en diferencias no lineal, y además el método es implícito hemos de resolver una ecuación en cada paso, para ello haremos uso de la librería *sympy*.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve

def MET_T1_EJ4_3(rango,n):
    a = rango[0] ; b = rango[1]
    h = (b-a)/n;
    S = np.zeros((n+1,2));
    S[0,:] = [0,1] #inicializamos con la c.i (x0,y0)
    S[1,:] = [h,(1+h**2)**2]; #inicializamos con la sol. exacta (x1,y1)
    y=symbols('y)')
    for i in range(1,n+1):#S[i:0]
        y2=solve(y-S[i-1,1]-(h/12)*(4*4*i*h*y**(1/2)+8*4*(i-1)*h*S[i-1,1]**(1/2)-4*(i-2)*h*S[i-2,1]**(1/2)))
        S[i,:]= [S[i-1,0]+h, y2[0]]
    return S
```

Calculamos ahora las soluciones para los h específicados, y creamos  una lista con soluciones ('sol'):

```python
n=[20,40,80]
sol = [];
for i in range(0,3): 
    sol.append(MET_T1_EJ4_3((0,2),n[i]))
print(sol)
```

La representación gráfica nos muestra las tres soluciones numéricas. 

```python
import matplotlib.pyplot as plt
plt.plot(sol[0][:, 0],sol[0][:, 1],'gs',sol[1][:, 0], sol[1][:, 1],'bd', sol[2][:, 0], sol[2][:, 1],'r-')
plt.legend(("h=0.1","h=0.05","h=0.025"))
```

```python
import matplotlib.pyplot as plt
plt.plot(sol[0][:, 0],sol[0][:, 1],'gs',sol[1][:, 0], sol[1][:, 1],'bd', sol[2][:, 0], sol[2][:, 1],'r-')
plt.legend(("h=0.1","h=0.05","h=0.025"))
plt.yscale('log')
```

 Si bien las tres soluciones se superponen y son practicamente indistinguibles a la vista, aún usando escala logaritmica. Por lo que vamos mostrar la información de las soluciuones de forma tabulada, mediante la impresión con formato.

## Ejercicio 5.3 

Demostrar el efecto de la cero-inestabilidad usando el método
$$ y_{n+2}-(1+a)y_{n+1}+ay_n=\frac{h}{2}\big[(3-a)f_{n+1}-(1+a)f_n \big]$$
con (i) $a=0$, (ii) $a=-5$ para calcular soluciones numéricas del PVI $y’= 4xy^{1/2}, y(0)=1$, para   $0 < x <2$, con una longitud de paso $h=0.1$, $0.05$ y $0.025$.  

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve

def MET_T1_EJ5_3(rango,n,a1):
    a = rango[0] ; b = rango[1]
    h = (b-a)/n;
    S = np.zeros((n+1,2));
    S[0,:] = [0,1] #inicializamos con la c.i (x0,y0)
    S[1,:] = [h,(1+h**2)**2]; #inicializamos con la sol. exacta (x1,y1)
    for i in range(2,n+1):
        y2=(1+a1)*S[i-1,1]-a1*S[i-2,1]+(h/2)*((3-a1)*4*(i-1)*h*S[i-1,1]**(1/2)-(1+a1)*4*(i-2)*h*S[i-2,1]**(1/2))
        S[i,:]= [S[i-1,0]+h, y2]
    return S
```
